```python
import pandas as pd

# Sample Data Preparation
final_cleaned_df['Signup_Date'] = pd.to_datetime(final_cleaned_df['Signup_Date'])
final_cleaned_df['System_First_Bet_Datetime'] = pd.to_datetime(final_cleaned_df['System_First_Bet_Datetime'])
final_cleaned_df['Date'] = pd.to_datetime('today')  # Use today's date for calculation

# Create cohorts based on signup month
final_cleaned_df['Signup_Month'] = final_cleaned_df['Signup_Date'].dt.to_period('M')

# Define a function to calculate retention rate
def calculate_retention_rates(df, period_days=30):
    df['Last_Active_Date'] = df[['System_First_Bet_Datetime', 'Date']].max(axis=1)
    df['Days_Active'] = (df['Date'] - df['Last_Active_Date']).dt.days
    
    retention_data = []
    for cohort in df['Signup_Month'].unique():
        cohort_data = df[df['Signup_Month'] == cohort]
        total_users = len(cohort_data)
        for days in [period_days]:
            retained_users = len(cohort_data[cohort_data['Days_Active'] <= days])
            retention_rate = (retained_users / total_users) * 100
            retention_data.append({
                'Cohort': cohort,
                'Period_Days': days,
                'Total_Users': total_users,
                'Retained_Users': retained_users,
                'Retention_Rate': retention_rate
            })
    return pd.DataFrame(retention_data)

# Calculate retention rates
retention_df = calculate_retention_rates(final_cleaned_df)

# Calculate churn rate
def calculate_churn_rate(df, period_days=30):
    df['Last_Active_Date'] = df[['System_First_Bet_Datetime', 'Date']].max(axis=1)
    df['Days_Active'] = (df['Date'] - df['Last_Active_Date']).dt.days
    
    churn_data = []
    for cohort in df['Signup_Month'].unique():
        cohort_data = df[df['Signup_Month'] == cohort]
        total_users = len(cohort_data)
        for days in [period_days]:
            churned_users = len(cohort_data[cohort_data['Days_Active'] > days])
            churn_rate = (churned_users / total_users) * 100
            churn_data.append({
                'Cohort': cohort,
                'Period_Days': days,
                'Total_Users': total_users,
                'Churned_Users': churned_users,
                'Churn_Rate': churn_rate
            })
    return pd.DataFrame(churn_data)

# Calculate churn rates
churn_df = calculate_churn_rate(final_cleaned_df)

# Display results
print("Retention Rates:")
print(retention_df)

print("\nChurn Rates:")
print(churn_df)
```
